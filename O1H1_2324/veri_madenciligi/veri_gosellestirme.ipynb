```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
veri = pd.read_csv("kalp_rahatsizligi.csv")
veri.head(5)
#comma seperate value
```

```python
veri["yas"]
```

```python
veri.yas
```

```python
#histogram grafiği sayısal verilerin dağılımını görselleştirir.
sns.histplot(data=veri,x="yas",bins=12)
plt.show()
```

```python
veri.columns
```

```python
#serum kolestrol histogramı
sns.histplot(data=veri,x="serum_kolestrol",bins=10)
plt.title("serum kolestrol histogramı")
plt.ylabel("Adet")
plt.show()
```

```python
#serum kolestrol histogramının kalp rahatsizliği verisine göre ayrıştırılması

sns.histplot(data=veri,x="serum_kolestrol",hue="kalp_rahatsizligi",element="step",bins=15)
plt.show()
```

```python
veri.columns
```

```python
# en_yuksek_kalp_hizi sayısal verisinin cinsiyet kategorik verisine göre histogram grafiğini çizdiriniz.
sns.histplot(data=veri,x='en_yuksek_kalp_hizi',hue="cinsiyet",element="step",bins=12)
plt.title("En yüksek kalp hızı/cinsiyet histogramı")
plt.show()
```

```python
#saçılım(scatter) grafiği iki sayısal veri arasındaki ilişkiyi ifade eder
sns.scatterplot(data=veri,x="yas",y="en_yuksek_kalp_hizi")
plt.title("yaş/en yükske kalp hızı saçılım grafiği")
plt.show()
```

```python
#saçılım(scatter) grafiği iki sayısal veri arasındaki ilişkiyi ifade eder
sns.lmplot(data=veri,x="yas",y="en_yuksek_kalp_hizi")
plt.title("yaş/en yükske kalp hızı saçılım grafiği")
plt.show()
```

```python
sns.lmplot(data=veri,x="en_yuksek_kalp_hizi",y="serum_kolestrol")
plt.show()
```

```python
#saçılım(scatter) grafiği iki sayısal veri arasındaki ilişkiyi ifade eder
sns.lmplot(data=veri,x="yas",y="en_yuksek_kalp_hizi",hue="cinsiyet")
plt.title("yaş/en yükske kalp hızı saçılım grafiği")
plt.show()
```

```python
#sütun garfiği kategorik verilerin dağılımı ile graikleri oluşturur.
g=sns.countplot(data=veri,x="cinsiyet")
g.bar_label(g.containers[0])
plt.show()
```

```python
#sütun grafiği kategorik verilerin dağılımı ile graikleri oluşturur.
sns.countplot(data=veri,x="cinsiyet",hue="kalp_rahatsizligi")
plt.show()
```

```python
sns.countplot(data=veri,x="gogus_agrisi_tipi",hue="kalp_rahatsizligi")
plt.show()
```

```python
#kutu(boxplot) grafiği sayısal verilerin istatistiksel olarak incelenmesini sağlayan garfiklerdir.

sns.boxplot(data=veri,y="serum_kolestrol")
plt.show()
```
