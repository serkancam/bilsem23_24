```python
from sklearn.model_selection import train_test_split

X=veri.iloc[:,:-1].values
y=veri.iloc[:,-1].values

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=22)
```

```python
from sklearn.linear_model import LinearRegression

regresyoncu = LinearRegression()

regresyoncu.fit(X_train,y_train)
```

```python

tahmin=regresyoncu.predict(  [[37,75],[70,30]]   )
print("tahmin edilen:",tahmin)
```

```python
egim=regresyoncu.coef_
kesisim=regresyoncu.intercept_
print(egim, kesisim)
```
