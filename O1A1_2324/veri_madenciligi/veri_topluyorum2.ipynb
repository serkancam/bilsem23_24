```python
import selenium.webdriver.firefox.service
from selenium import webdriver
from bs4 import BeautifulSoup as bs4
import os
import time


# temp yolu oluştur
temp_dir = "~/_tmp"
if not os.path.exists(temp_dir):
    os.makedirs(temp_dir)
os.environ["TMPDIR"] = temp_dir

# firefox yolunu göster
opt = webdriver.FirefoxOptions()
opt.binary_location = "/usr/bin/firefox"
# opt.add_argument("--headless")

# driver yolunu göster
srv = selenium.webdriver.firefox.service.Service()
srv.path = "/home/serkan/Belgeler/bilsem23_24/O1A1_2324/veri_madenciligi/geckodriver"

driver = webdriver.Firefox(options=opt, service=srv)

driver.get("https://www.epey.com/laptop")

tarih=[]



# print(icerik)
for i in range(5):
    icerik = driver.page_source
    try:
        ayristirma_agaci=bs4(icerik,"lxml")
        urun_listesi=ayristirma_agaci.find_all("ul",attrs={"class":"metin row"})
        # print(urun_listesi)
        for urun in urun_listesi:
            # print(urun.find_all("li")[2])
            fiyat=str(urun.find_all("li")[1].find("a").text)
            fiyat=fiyat.split("TL")[0]
            isim=str(urun.find_all("li")[1].find("a").get("title"))
            print(isim,"---",fiyat)
            
        driver.find_element_by_xpath("/html/body/div[1]/div/div[4]/div[2]/div[2]/div/div[3]/a[15]").click()
        time.sleep(2)
            
    except Exception as ex:
        print(ex)
    else:
        print("hatasız geçiş")

driver.close()
```
