```python
# veri_gorsellestirme.ipynb
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
veri=pd.read_csv("kalp_rahatsizligi.csv")
veri.head(5)
```

```python
veri.columns
```

```python
veri.info()
```

```python
sns.histplot(x="yas",data=veri,bins=15)
plt.xlabel("Yaş")
plt.title("Yaş histogramı")
plt.ylabel("Sayı")
plt.show()
# histogram sayısal veriler için kullanılır.
```

```python
#hareketsiz_kan_basinci verisine ait histogram grafiğini çizdiriniz.
sns.histplot(x="hareketsiz_kan_basinci",data=veri,bins=15)
plt.xlabel("hareketsiz_kan_basinci")
plt.title("hareketsiz_kan_basinci histogramı")
plt.ylabel("Sayı")
plt.show()
```

```python
sns.histplot(x="yas",data=veri,hue="kalp_rahatsizligi",element="step",bins=10)
plt.ylabel("Kişi sayısı")
plt.title('"Yaş verisinin kalp rahatsızlığı verisine göre dağılımı"')
plt.show()
```

```python
sns.histplot(x="hareketsiz_kan_basinci",data=veri,hue="kalp_rahatsizligi",element="step",bins=10)
plt.ylabel("Kişi sayısı")
plt.title('"Kan basıncı verisinin kalp rahatsızlığı verisine göre dağılımı"')
plt.show()
```

```python
#saçılım(scatter) grafikleri
# iki sayısal veri arasındaki ilişkişi gözlemlemek için kullanılır.
#örneğin yaş ve en yüksek kalp hızı

sns.scatterplot(data=veri, x="yas",y="en_yuksek_kalp_hizi")
sns.lmplot(data=veri, x="yas",y="en_yuksek_kalp_hizi")
plt.ylabel("kalp hızı")
plt.xlabel("yaş")
plt.title("Yaş/En yüksek kalp hızı saçılım")
plt.show()
```

```python
#en_yuksek_kalp_hizi  verisi ile hareketsiz_kan_basinci verisinin saçılım grafiğini oluşturunuz
sns.lmplot(data=veri, x="en_yuksek_kalp_hizi",y="hareketsiz_kan_basinci")

plt.ylabel("hareketsiz_kan_basinci")
plt.xlabel("en_yuksek_kalp_hizi")
plt.title("Yaş/En yüksek kalp hızı saçılım")
plt.show()
```

```python
#kutu grafikleri
#sayısal veriler içindir.

sns.boxplot(data=veri,y="serum_kolestrol")
plt.ylabel("serum kolestrol")
plt.title("Serum koelstrol kutu grafiği")
plt.show()
veri.columns
```

```python
sns.boxplot(data=veri,y="serum_kolestrol",x="cinsiyet")
plt.ylabel("serum kolestrol")
plt.title("Cinsiyete göre serum kolestrol kutu grafiği")
plt.show()
```

```python
#kategorik verilerin sayısını görmek için sütun grafikleri kullanılır.
g=sns.countplot(data=veri,x="cinsiyet")
plt.xlabel("cinsiyet")
plt.ylabel("adet")
plt.title("Kadın/Erkek sayıları")
g.bar_label(g.containers[0])
plt.show()

```
