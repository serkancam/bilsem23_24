```python
# veri_gorsellestirme.ipynb
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#veri setinin pandas ile okunmasÄ±
veri = pd.read_csv("kalp_rahatsizligi.csv")
veri.head(3)
```

```python
# print(type(veri))
veri.info()
```

```python
veri.describe()
```

```python
#histogram grafiÄi: sayÄ±sal verilerin daÄÄ±lÄ±mÄ±nÄ± gÃ¶sterir.

sns.histplot(data=veri,x="yas",bins=12)
plt.xlabel("YaÅ")
plt.ylabel("Adet")
plt.title("YaÅ verisi histogramÄ±")
# plt.yticks(range(0,50,2))
# plt.xticks(range(20,80,2))
plt.show()
```

```python
ax = sns.histplot(data=veri,x="yas", bins=12) 
ax.bar_label(ax.containers[0])
ax.set_yticks(ticks=range(0,50,5),minor=True)
plt.show()
```

```python
veri.columns
```

```python
# en yÃ¼ksek kalp hÄ±zÄ± verisinin histogram grafiÄini oluÅturan kodu yazÄ±nÄ±z.
sns.histplot(data=veri,x="en_yuksek_kalp_hizi",bins=15)
plt.xlabel("en yÃ¼ksek kalp hÄ±zÄ±")
plt.ylabel("Adet")
plt.show()
```

```python
# sayÄ±sal verileri kategorik veriler ile ayrÄ±ÅtÄ±rÄ±p o Åekilde Ã§izim yapabiliriz.
# en yÃ¼ksek kalp hÄ±zÄ± verisinin histogram grafiÄini oluÅturan kodu yazÄ±nÄ±z.


sns.histplot(data=veri,x="en_yuksek_kalp_hizi",bins=10,hue="cinsiyet",element="step")
plt.xlabel("en yÃ¼ksek kalp hÄ±zÄ±")
plt.ylabel("Adet")
plt.show()
```

```python
# en yÃ¼ksek kalp hÄ±zÄ± verisini kalp rahatsÄ±zlÄ±ÄÄ± verisine gÃ¶re daÄlÄ±mÄ±nÄ± Ã§izdiriniz.
sns.histplot(data=veri,x="en_yuksek_kalp_hizi",hue="kalp_rahatsizligi",element="step")
plt.legend(["yok","var"],loc="upper left")
plt.show()
```

```python
# sÃ¼tun grafiÄi kategorik verilerin adetlerini gÃ¶rmek iÃ§in kullanÄ±lÄ±r

grafik=sns.countplot(data=veri,x="cinsiyet")
grafik.bar_label(grafik.containers[0])
plt.show()
```

```python
# sÃ¼tun grafiÄi kategorik verilerin adetlerini gÃ¶rmek iÃ§in kullanÄ±lÄ±r

sns.countplot(data=veri,x="cinsiyet")

plt.show()
```

```python
g=sns.countplot(data=veri,x="kalp_rahatsizligi")
g.bar_label(g.containers[0])
plt.xticks(ticks=[0,1],labels=["yok","var"])
plt.show()
```

```python
# kadÄ±n ve erkekler arasÄ±ndaki kalp rahatszÄ±lÄ±ÄÄ± sayÄ±sÄ±

sns.countplot(x="cinsiyet",hue="kalp_rahatsizligi",data=veri)
plt.show()
```

```python
# aÃ§lÄ±k kan Åekeri ve kalp_rahatsÄ±zlÄ±ÄÄ±
sns.countplot(x="aclik_kan_sekeri",hue="kalp_rahatsizligi",data=veri)
plt.show()
```
