```python
# veri_gorsellestirme.ipynb
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#veri setinin pandas ile okunması
veri = pd.read_csv("kalp_rahatsizligi.csv")
veri.head(3)
```

```python
# print(type(veri))
veri.info()
```

```python
veri.describe()
```

```python
#histogram grafiği: sayısal verilerin dağılımını gösterir.

sns.histplot(data=veri,x="yas",bins=12)
plt.xlabel("Yaş")
plt.ylabel("Adet")
plt.title("Yaş verisi histogramı")
# plt.yticks(range(0,50,2))
# plt.xticks(range(20,80,2))
plt.show()
```

```python
ax = sns.histplot(data=veri,x="yas", bins=12) 
ax.bar_label(ax.containers[0])
ax.set_yticks(ticks=range(0,50,5),minor=True)
plt.show()
```

```python
veri.columns
```

```python
# en yüksek kalp hızı verisinin histogram grafiğini oluşturan kodu yazınız.
sns.histplot(data=veri,x="en_yuksek_kalp_hizi",bins=15)
plt.xlabel("en yüksek kalp hızı")
plt.ylabel("Adet")
plt.show()
```

```python
# sayısal verileri kategorik veriler ile ayrıştırıp o şekilde çizim yapabiliriz.
# en yüksek kalp hızı verisinin histogram grafiğini oluşturan kodu yazınız.


sns.histplot(data=veri,x="en_yuksek_kalp_hizi",bins=10,hue="cinsiyet",element="step")
plt.xlabel("en yüksek kalp hızı")
plt.ylabel("Adet")
plt.show()
```

```python
# en yüksek kalp hızı verisini kalp rahatsızlığı verisine göre dağlımını çizdiriniz.
sns.histplot(data=veri,x="en_yuksek_kalp_hizi",hue="kalp_rahatsizligi",element="step")
plt.legend(["yok","var"],loc="upper left")
plt.show()
```

```python
# sütun grafiği kategorik verilerin adetlerini görmek için kullanılır

grafik=sns.countplot(data=veri,x="cinsiyet")
grafik.bar_label(grafik.containers[0])
plt.show()
```

```python
# sütun grafiği kategorik verilerin adetlerini görmek için kullanılır

sns.countplot(data=veri,x="cinsiyet")

plt.show()
```

```python
g=sns.countplot(data=veri,x="kalp_rahatsizligi")
g.bar_label(g.containers[0])
plt.xticks(ticks=[0,1],labels=["yok","var"])
plt.show()
```

```python
# kadın ve erkekler arasındaki kalp rahatszılığı sayısı

sns.countplot(x="cinsiyet",hue="kalp_rahatsizligi",data=veri)
plt.show()
```

```python
# açlık kan şekeri ve kalp_rahatsızlığı
sns.countplot(x="aclik_kan_sekeri",hue="kalp_rahatsizligi",data=veri)
plt.show()
```

```python
# saçılım grafikleri(scatterplot) iki sayısal veri arasındaki ilişkiyi ortaya koyar.

sns.scatterplot(data=veri,x="yas",y="en_yuksek_kalp_hizi")
plt.show()

#KORELASYON(correlation)
```

```python
# saçılım grafikleri(scatterplot) iki sayısal veri arasındaki ilişkiyi ortaya koyar.
sns.lmplot(data=veri,x="yas",y="en_yuksek_kalp_hizi",hue="cinsiyet")

plt.show()
```
