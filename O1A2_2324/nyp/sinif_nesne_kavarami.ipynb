### Sınıf kavramı:
sınıf,  bir varlığın program içerisinde özellik ve davranışlarını modelleyebileceğimiz bir tasarım şablonudur.


### Nesne kavramı:

nesne, sınıf ile tasarlanmış varlığın, hayat bulmuş halidir. Yani bir sınıf tanımlamasından nesne türetilmemişse o sınıf kullanılmamış demektir.


```python
# bir sınıf tanımlayalım

class oyuncu():
    def __init__(self) :
        self.adi=""
        self.gucu=0.0
        self.savunma=0.0
        self.saldiri_gucu=0.0
        self.takim=0
        

a=oyuncu() # oyuncu sınıfının yapıcı methodu ile  yeni nesne a oluşturuluyor.
b=oyuncu() # oyuncu sınıfının yapıcı methodu ile  yeni nesne b oluşturuluyor.
a.adi="a karakteri"
b.adi="b karakteri"
a.gucu=100
b.gucu=98


print(a.adi,a.gucu)
print(b.adi,b.gucu)



# 10 tane oyuncu sınıfından nesne tanımlayıp listeye ekleyelim

oyuncular=list()

for _ in range(10):
    oyuncular.append(oyuncu())
```

* Eğer bir sınıf tanımında __init__ yapıcı methodu yoksa varsayılan yapıcı method kullanılır.
* Bir sınıftan nesne türetilirken aslında olan durum, sınıfın yapıcı methodu(__init__) çalışarak sınıfın bir kopyasını ilgili konuma aktarır. 
* sinif_adi() yazım şeklinde çalışan method nesne türetme işini yapar.
