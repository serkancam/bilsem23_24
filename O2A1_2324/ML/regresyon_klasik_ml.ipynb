```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# veri dosyadan okundu ve veri_ham değişkenine aktarıldı.
veri_ham = pd.read_csv("regresyon_veri2.csv")
```

```python
# dosyanın temiz hali veri değişkenine aktarıldı
veri = veri_ham.copy()
```

```python
# verinin ilk 5 satırnını görelim
veri.head(5)
```

```python
veri
```

```python
X=veri.iloc[:,:-1]#virgülün solundaki : tüm satırları dahil et, virgülün sağındaki :-1 son sütun hariç  tüm sütunları veriye dahil et demek:
print(type(X))# bu veri pandas
print(type(X.values)) # bu veri numpy ndarray

# makine öğrenmesi kütüphaneleri(sklearn, tensorflow, keras, pytorch ....) verileri numpy yapısında isterler 
X=X.values
print(type(X))
```

```python
X
```

```python
y=veri.iloc[:,-1].values
y
```

```python
from sklearn.linear_model import LinearRegression

regresyoncu = LinearRegression()

regresyoncu.fit(X,y) # X lerden y leri elde eden ax1+bx2+c denklemindeki a ve b eğim değerleri ile c kesim değerini bul
```

```python
egimler=regresyoncu.coef_ 
kesisim=regresyoncu.intercept_
```

```python
print(egimler,kesisim)
```
