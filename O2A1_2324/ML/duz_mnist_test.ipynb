```python
# duz_mnist_test.ipynb
import keras 
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
model=keras.models.load_model("/home/serkan/Belgeler/bilsem23_24/O2A1_2324/ML/duz_minst.h5")
```

```python
resim = cv2.imread("r42.png")
plt.imshow(resim)
plt.show()
print(resim.shape)
#gri tonlamaya çevir
resim=cv2.cvtColor(resim,cv2.COLOR_BGR2GRAY)
print(resim.shape)
# 28*28 piksel boyutuna getir
resim=cv2.resize(resim,(28,28))

print(resim.shape)
#resim tahmin işlemine hazır
resim=resim.reshape(1,28,28)#unutma

tahmin=model.predict(resim.astype("float32")/255)
print("tahmin",np.argmax(tahmin))

```
