```python
import keras
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
model = keras.models.load_model("/home/serkan/Belgeler/bilsem23_24/O2H1_2324/ML/mnist_cnn.h5")
```

```python
#resmi alma
resim = cv2.imread("r9.png")
resim = cv2.cvtColor(resim,cv2.COLOR_BGR2GRAY)
resim=cv2.resize(resim,(28,28))
print(resim.shape)
plt.imshow(resim,cmap="gray")
plt.show()
resim=resim.astype("float32")/255
resim=resim.reshape(1,28,28,1)
tahmin=model.predict(resim)
print(tahmin)
print(np.argmax(tahmin))
```
