```python
import pandas as pd
import numpy as np
```

```python
veri=pd.read_csv("regresyon_veri2.csv")
veri.head(10)
```

```python
# -1,+1 arasına pandas ile normalize etmek
veri_ham = veri.copy()
ort = veri.mean()
std = veri.std()
veri=(veri-ort)/std
veri.head(5)
```

```python
# -1,1 aralığına sklearn ile normalize
from sklearn.preprocessing import minmax_scale
X1= minmax_scale(veri_ham.iloc[:,:-1].values,feature_range=(-1,1))
X1[0:5]
```

```python
X= veri.iloc[:,:-1].values
y=veri.iloc[:,-1].values
```

```python
import tensorflow as tf
from keras import layers,models
```

```python
ysa_model = models.Sequential() # sıralı model
# şimdi bu modele düz(Dense) katmanları ekleyelim.

# giriş katmanı kendinden sonraki ilk katman ile beraber eklenir.
ysa_model.add(layers.Dense(16,input_shape=(2,),activation="relu"))
ysa_model.add(layers.Dense(8,activation="relu"))
ysa_model.add(layers.Dense(1))
ysa_model.compile(loss="mse",optimizer="rmsprop",metrics=["mae"])
print(ysa_model.summary())
```

```python
egitim_hikayesi = ysa_model.fit(X,y,epochs=1000,validation_split=0.2)
```

```python
test_veri=np.array([[40,50],[7,4]])

test_veri=(test_veri-ort.values[0:2])/std.values[0:2]

tahmin = ysa_model.predict(test_veri)
print("normalize öncesi",tahmin)
tahmin=tahmin*std.values[-1]+ort.values[-1]
print("normalzie sonrası",tahmin)
```
